# Seaborn Practice

1.Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.
2.used for data visualization and based on Matplotlib.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a=sns.load_dataset('flights')
```

```python
a.head()
```

```python
sns.relplot(x='passengers',y='month',data=a)
```

# To add another dimension

```python
a=sns.load_dataset('flights')
```

```python
sns.relplot(x='passengers',y='month',z='year',data=a)
```

# For line plot

```python
b=sns.load_dataset('tips')
```

```python
sns.relplot(x='time',y='tip',data=b,kind='line')
```

# Plotting with categorical data

```python
b=sns.load_dataset('tips')
```

```python
sns.catplot(x='day',y='total_bill',data=b)
```

# For violine plot

```python
sns.catplot(x='day',y='total_bill',kind='violin',data=b)
```

# For boxen plot

```python
sns.catplot(x='day',y='total_bill',kind='boxen',data=b)
```

# Plot Aesthestics

Deals with making our plots more attractive and delightful using various themes and color-pallettes

Lets try::

```python
sns.set(style='darkgrid')
a=sns.load_dataset('flights')
b=sns.PairGrid(a)
b.map(plt.scatter)
```

```python
sns.set(style='white',color_codes=True)
a=sns.load_dataset('tips')
sns.boxplot(x='day',y='total_bill',data=a)
```

Want to remove Piece axis

use despine function

```python
sns.set(style='white',color_codes=True)
a=sns.load_dataset('tips')
sns.boxplot(x='day',y='total_bill',data=a)
sns.despine(offset=10,trim=True)
```
